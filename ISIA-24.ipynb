```python
import warnings
import os
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.metrics import classification_report, precision_recall_curve, roc_curve, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Suppress warnings
warnings.filterwarnings("ignore")

# Define the model names
model_names = [
    'aubmindlab/bert-base-arabertv02',
    'aubmindlab/bert-large-arabertv02',
    'aubmindlab/araelectra-base-discriminator',
    'alger-ia/dziribert_sentiment'
]

# Define datasets
datasets = {
    "arabic_original": "/kaggle/input/arabic-ml/arabic_V3.csv",
    "mask_arabic_V3": "/kaggle/input/arabic-ml/mask_arabic_V3.csv",
    "random_deletion_augmented_arabic_V3": "/kaggle/input/arabic-ml/random_deletion_augmented_arabic_V3.csv",
    "random_insertion_augmented_arabic_V3": "/kaggle/input/arabic-ml/random_insertion_augmented_arabic_V3.csv",
    "shufell_arabic_V3": "/kaggle/input/arabic-ml/shufell_arabic_V3.csv",
    "n_gram_insertion": "/kaggle/input/augmentatio-random-inseration/arabic_V3 - arabic_V3.csv"
}

# Function to extract losses from log history
def extract_losses(log_history):
    train_losses = []
    val_losses = []
    for log in log_history:
        if 'loss' in log:
            train_losses.append(log['loss'])
        if 'eval_loss' in log:
            val_losses.append(log['eval_loss'])
    return {'train': train_losses, 'val': val_losses}

# Function to plot PRC and ROC curves
def plot_curves(model_name, curves_data):
    # Plot Precision-Recall Curve
    plt.figure(figsize=(12, 6))
    for recall, precision, dataset_name in curves_data['prc']:
        plt.plot(recall, precision, marker='.', label=f'{dataset_name}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve for {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot ROC Curve
    plt.figure(figsize=(12, 6))
    for fpr, tpr, dataset_name in curves_data['roc']:
        plt.plot(fpr, tpr, marker='.', label=f'{dataset_name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Process each model
for model_name in model_names:
    print(f"Processing model: {model_name}")

    # Load the pre-trained model and tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2, ignore_mismatched_sizes=True)

    # Freeze the base layer of the model
    for param in model.base_model.parameters():
        param.requires_grad = False

    # Initialize lists to store results
    original_dataset_results = {}
    augmented_dataset_results = {'prc': [], 'roc': []}

    # Process each dataset
    for dataset_name, dataset_path in datasets.items():
        print(f"Processing dataset: {dataset_name}")

        # Load the dataset
        data = pd.read_csv(dataset_path)
        data = data[['new_comments', 'omar annotation']]

        # Filter out neutral labels
        data = data[data['omar annotation'] != 'neutral']

        # Map the labels to numerical values
        label_mapping = {'positive': 1, 'negative': 0}
        data['label'] = data['omar annotation'].map(label_mapping)

        # Drop rows with non-finite values in the 'label' column
        data = data.dropna(subset=['label'])

        # Convert label column to integer type
        data['label'] = data['label'].astype(int)

        # Split the dataset into train and validation sets
        train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

        # Tokenize the data
        train_encodings = tokenizer(list(train_data['new_comments'].values), truncation=True, padding=True, return_tensors='pt')
        val_encodings = tokenizer(list(val_data['new_comments'].values), truncation=True, padding=True, return_tensors='pt')

        # Define the dataset class
        class ArabicDataset(torch.utils.data.Dataset):
            def __init__(self, encodings, labels):
                self.encodings = encodings
                self.labels = labels

            def __getitem__(self, idx):
                item = {'input_ids': self.encodings['input_ids'][idx],
                        'attention_mask': self.encodings['attention_mask'][idx]}
                item['labels'] = self.labels[idx]
                return item

            def __len__(self):
                return len(self.labels)

        train_dataset = ArabicDataset(train_encodings, train_data['label'].values)
        val_dataset = ArabicDataset(val_encodings, val_data['label'].values)

        # Define the training arguments
        training_args = TrainingArguments(
            output_dir='./results',
            save_strategy="no",
            num_train_epochs=5,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=8,
            gradient_accumulation_steps=4,
            evaluation_strategy="epoch",
            logging_dir='./logs',
            logging_steps=10,
            disable_tqdm=False,  # Enable tqdm progress bar
            do_train=True,
            report_to='none'  # Disable logging to prevent issues with metrics
        )

        # Create the Trainer object
        trainer = Trainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
        )

        # Fine-tune the model
        trainer.train()

        # Evaluate the model
        predictions = trainer.predict(val_dataset)
        predicted_labels = np.argmax(predictions.predictions, axis=1)
        true_labels = val_dataset[:]['labels']

        # Calculate metrics
        report = classification_report(true_labels, predicted_labels, target_names=['0', '1'], output_dict=True)

        # Extract losses from log history
        losses = extract_losses(trainer.state.log_history)

        # Calculate PRC and ROC curves
        prc = precision_recall_curve(true_labels, predicted_labels)
        roc = roc_curve(true_labels, predicted_labels)

        # Update results based on dataset
        if dataset_name == "arabic_original":
            original_dataset_results = {
                "accuracy": report['accuracy'],
                "macro_precision": report['macro avg']['precision'],
                "macro_recall": report['macro avg']['recall'],
                "macro_f1": report['macro avg']['f1-score'],
                "support": report['macro avg']['support'],
                "prc": (prc[0], prc[1]),
                "roc": (roc[0], roc[1]),
                "losses": losses
            }
        else:
            augmented_dataset_results['prc'].append((prc[0], prc[1], dataset_name))
            augmented_dataset_results['roc'].append((roc[0], roc[1], dataset_name))

        # Print the classification report
        print(f"Model: {model_name}")
        print(f"Dataset: {dataset_name}")
        print(classification_report(true_labels, predicted_labels, target_names=['0', '1']))

        # Plot confusion matrix
        cm = confusion_matrix(true_labels, predicted_labels)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['0', '1'])
        plt.figure()
        disp.plot()
        plt.title(f'Confusion Matrix: {dataset_name}')
        plt.show()

    # Plot PRC and ROC curves for the model
    plot_curves(
        model_name,
        {
            'prc': augmented_dataset_results['prc'] + [(original_dataset_results['prc'][0], original_dataset_results['prc'][1], 'Original Dataset')],
            'roc': augmented_dataset_results['roc'] + [(original_dataset_results['roc'][0], original_dataset_results['roc'][1], 'Original Dataset')]
        }
    )
```
